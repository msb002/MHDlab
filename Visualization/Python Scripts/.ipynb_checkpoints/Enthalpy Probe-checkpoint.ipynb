```python
import numpy as np
import pandas as pd
from nptdms import TdmsFile as TF
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline

import datetime as dt
```

```python
def maketime(string):
    mo = int(string[0:2])
    day = int(string[3:5])
    yr = int(string[6:10])
    hr = int(string[11:13])
    mn = int(string[14:16])
    sec = int(string[17:19])
    return dt.datetime(yr, mo, day, hr+7, mn, sec)
```

```python
file629 = TF("Z:\\Test Data\\Raw Data\\EnthalpyProbe\\2018-07-03\\Run 1_together.tdms")
jp = pd.read_excel("Z:\\Test Data\\Raw Data\\JP8000\\Reports\\PYROMANIAC2673.xls", sheetname = "HVOFGun")
```

```python
realtime = []
for time in jp['Time']:
    realtime.append(maketime(time))
jp['Time'] = realtime
```

```python
ff629 = file629.object("Untitled", "FuelFlow (Arith. Mean)").data
time1 = file629.object('Global', 'Time').data
zero1 = min(time1)
newtime1 = []
for time in time1:
    newtime1.append((time-zero1).seconds)
```

```python
plt.figure(figsize = (13,8))
plt.title('JP vs flow meter Fuel Flow')
plt.ylabel('Fuel Flow (L/min)')
plt.xlabel('Time')
plt.plot(time1, ff629, label='Flow Meter')
plt.plot(jp['Time']-dt.timedelta(seconds=74), jp['Fuel Flow']/60, label = 'JP')
plt.legend()
```

```python
of629 = file629.object("Untitled", "Mass Flow_O2").data
```

```python
plt.figure(figsize = (13,8))
plt.title('JP vs Alicat O2 Flow')
plt.ylabel('O2 (SLPM)')
plt.xlabel('Time')

plt.plot(time1, of629, label = 'Alicat')
plt.plot(jp['Time']-dt.timedelta(seconds=74), jp["Oxygen Flow"]*(293/298)*(14.696/14.504), label = 'JP')
plt.legend()
```

```python
vof = file629.object("Untitled", "Volumetric Flow_O2").data
ot = file629.object("Untitled", "Temperature_O2").data
op = file629.object("Untitled", "Absolute Pressure_O2").data
```

```python
mofcalc = (298.15/14.696)*vof*op/(ot+273.15)
```

```python
plt.plot(time1, mofcalc, label = 'User Calc. Mass Flow')
plt.plot(time1, of629, label = 'Cont. Calc. Mass Flow')
plt.legend()
plt.xlabel('Time')
plt.ylabel("Mass Flow (SLPM)")
plt.title('Comparison of O2 Mass Flow from Alicat')
#lt.plot(jp['Time'],jp['Oxygen Flow'])
```

```python
stpof = (273.15/14.696)*vof*op/(ot+273.15)*(32/22.4)*(1/60)
```

```python
plt.figure(figsize = (13,8))
#plt.axvline(time1[10000])
plt.plot(time1, stpof)
plt.xlabel('Time')
plt.ylabel("Mass Flow (g/s)")
plt.title('Alicat O2 Mass Flow (g/s)')
```

```python
plt.figure(figsize = (13,8))
plt.xlabel('Time')
plt.ylabel("Mass Flow (g/s)")
plt.title('Fuel Mass Flow (g/s)')
plt.plot(time1, ff629*1000/60)
plt.plot(jp["Time"]-dt.timedelta(seconds=74), jp["Fuel Flow"]*1000/3600)
```
